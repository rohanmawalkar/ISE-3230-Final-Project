```python
import gurobipy as gp
from gurobipy import GRB

m = gp.Model("age_of_empires_ii")
m.Params.LogToConsole = 0 

s = m.addVars(50, name="s", vtype=GRB.BINARY) # whether a stone building was made during the turn
f = m.addVars(50, name="f", vtype=GRB.BINARY) # whether a food building was made during the turn
w = m.addVars(50, name="w", vtype=GRB.BINARY) # whether a wood building was made during the turn
g = m.addVars(50, name="g", vtype=GRB.BINARY) # whether a gold building was made during the turn

p = m.addVars(50, lb=0, name="w", vtype=GRB.INTEGER) # population made in each turn

sol1 = m.addVars(50, lb=0, name="sol1", vtype=GRB.INTEGER) # amt of soldier type 1 made in each turn
sol2 = m.addVars(50, lb=0, name="sol2", vtype=GRB.INTEGER) # amt of soldier type 2 made in each turn

sp = m.addVars(50, name="sp", vtype=GRB.INTEGER) # num population assigned to stone buildings during this turn
fp = m.addVars(50, name="fp", vtype=GRB.INTEGER) # num population assigned to food buildings during this turn
wp = m.addVars(50, name="wp", vtype=GRB.INTEGER) # num population assigned to wood buildings during this turn
gp = m.addVars(50, name="gp", vtype=GRB.INTEGER) # num population assigned to gold buildings during this turn



#m.setObjective(, GRB.MAXIMIZE) #TODO:add objective function


m.optimize()

print("obj_func = ", m.objVal)
for v in m.getVars():
    print('%s = %g' % (v.varName, v.x))
```
